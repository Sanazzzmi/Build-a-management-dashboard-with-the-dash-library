```python
import dash_bootstrap_components as dbc
from dash import Dash,dcc,html,Input,Output
import plotly.express as px
#!pip install jupyter-dash
from jupyter_dash import JupyterDash
import numpy as np
import pandas as pd
data = pd.read_csv('avocado.csv')
data["Date"] = pd.to_datetime(data["Date"], format="%Y-%m-%d")
data.sort_values("Date", inplace=True)

data
```

```python
external_stylesheets = [
    {
        "href": "https://fonts.google.com/css2?"
        "family=Lato:wght@400;700&display=swap",
        "rel": "stylesheet",
    },
]

app = JupyterDash(__name__, external_stylesheets=external_stylesheets)

app.title = "Avocado Analytics: Understand Your Avocados!"

app.layout = html.Div(
    children=[
        html.Div(
            children=[
                html.H1(
                    children="Avocado Analytics", className="header-title"
                ),
                html.P(
                    children="Analyze the behavior of avocado prices"
                    " and the number of avocados sold in the US"
                    " between 2015 and 2018",
                    className="header-description",
                ),
            ],
            className="header",
        ),
        
        #_______________first menu
        
        html.Div(
            children=[
                html.Div(
                    children=[
                        html.Div(children="Region", className="menu-title"),
                        dcc.Dropdown(
                            id="region-filter",
                            options=[
                                {"label": region, "value": region}
                                for region in np.sort(data.region.unique())
                            ],
                            value="Albany",
                            clearable=False,
                            className="dropdown",
                        ),
                    ]
                ),
                html.Div(
                    children=[
                        html.Div(children="Type", className="menu-title"),
                        dcc.Dropdown(
                            id="type-filter",
                            options=[
                                {"label": avocado_type, "value": avocado_type}
                                for avocado_type in data.type.unique()
                            ],
                            value="organic",
                            clearable=False,
                            searchable=False,
                            className="dropdown",
                        ),
                    ],
                ),
                html.Div(
                    children=[
                        html.Div(
                            children="Date Range",
                            className="menu-title"
                            ),
                        dcc.DatePickerRange(
                            id="date-range",
                            min_date_allowed=data.Date.min().date(),
                            max_date_allowed=data.Date.max().date(),
                            start_date=data.Date.min().date(),
                            end_date=data.Date.max().date(),
                        ),
                    ]
                ),
            ],
            className="menu",
        ),
        
        #_______________first figure
        
        html.Div(
            children=[
                html.Div(
                    children=dcc.Graph(
                        id="price-chart", config={"displayModeBar": False},
                    ),
                    className="card",
                ),
                html.Div(
                    children=dcc.Graph(
                        id="volume-chart", config={"displayModeBar": False},
                    ),
                    className="card",
                ),
            ],
            className="wrapper",
        ),
        
        
        #_______________second menu
        
        html.Div(
            children=[
                html.Div(
                    children=[
                        html.Div(children="Plu", className="menu-title"),
                        dcc.Dropdown(
                            id="plu-filter",
                            options=[
                                {"label": region, "value": region}
                                for region in data.columns[4:7]
                            ],
                            value=data.columns[4],
                            clearable=False,
                            className="dropdown",
                        ),
                    ]
                ),
                html.Div(
                    children=[
                        html.Div(children="bag", className="menu-title"),
                        dcc.Dropdown(
                            id="bag-filter",
                            options=[
                                {"label": avocado_type, "value": avocado_type}
                                for avocado_type in data.columns[8:11]
                            ],
                            value=data.columns[8],
                            clearable=False,
                            searchable=False,
                            className="dropdown",
                        ),
                    ],
                ),
            ],
            className="menu",style = {'margin' : '+10px auto 0 auto'}
        ),
        
        #_______________second figure
        
        html.Div(
            children=[
                html.Div(
                    children=dcc.Graph(
                        id="plu-chart", config={"displayModeBar": False},
                    ),
                    className="card",
                ),
                html.Div(
                    children=dcc.Graph(
                        id="bag-chart", config={"displayModeBar": False},
                    ),
                    className="card",
                ),
            ],className="wrapper",
        )
           
    ]
)


@app.callback(
    [
        Output("price-chart", "figure"),
        Output("volume-chart", "figure"),
        Output("plu-chart", "figure"),
        Output("bag-chart", "figure"),
     
    ],
    
    [
        Input("region-filter", "value"),
        Input("type-filter", "value"),
        Input("date-range", "start_date"),
        Input("date-range", "end_date"),
        Input("plu-filter", "value"),
        Input("bag-filter", "value"),
    ],
)


def update_charts(region, avocado_type, start_date, end_date,plu, bag):
    mask = (
        (data.region == region)
        & (data.type == avocado_type)
        & (data.Date >= start_date)
        & (data.Date <= end_date)
    )
    filtered_data = data.loc[mask, :]
    price_chart_figure = {
        "data": [
            {
                "x": filtered_data["Date"],
                "y": filtered_data["AveragePrice"],
                "type": "lines",
                "hovertemplate": "$%{y:.2f}<extra></extra>",
            },
        ],
        "layout": {
            "title": {
                "text": "Average Price of Avocados",
                "x": 0.05,
                "xanchor": "left",
            },
            "xaxis": {"fixedrange": True},
            "yaxis": {"tickprefix": "$", "fixedrange": True},
            "colorway": ["#17B897"],
        },
    }

    volume_chart_figure = {
        "data": [
            {
                "x": filtered_data["Date"],
                "y": filtered_data["Total Volume"],
                "type": "lines",
            },
        ],
        "layout": {
            "title": {"text": "Avocados Sold", "x": 0.05, "xanchor": "left"},
            "xaxis": {"fixedrange": True},
            "yaxis": {"fixedrange": True},
            "colorway": ["#E12D39"],
        },
    }
    
    plu_figure = {
        "data": [
            {
                "x": filtered_data["Date"],
                "y": data[plu],
                "type": "lines",
            },
        ],
        "layout": {
            "title": {"text": "Total number of avocados with PLU 4046 sold", "x": 0.05, "xanchor": "left"},
            "xaxis": {"fixedrange": True},
            "yaxis": {"fixedrange": True},
            "colorway": ["#17B897"],
        },
    }
    
    bag_figure = {
        "data": [
            {
                "x": filtered_data["Date"],
                "y": data[bag],
                "type": "lines",
            },
        ],
        "layout": {
            "title": {"text": "Total number of avocados with intended bag size sold", "x": 0.05, "xanchor": "left"},
            "xaxis": {"fixedrange": True},
            "yaxis": {"fixedrange": True},
            "colorway": ["#E12D39"],
        },
    }
    
    return price_chart_figure, volume_chart_figure , plu_figure , bag_figure


app.run_server()
```
